```python
import random
from datetime import datetime, timedelta
import pandas as pd
from pathlib import Path

random.seed(42)

out_path = Path("data/raw/labs_sample.csv")
out_path.parent.mkdir(parents=True, exist_ok=True)

patients = [f"P{str(i).zfill(3)}" for i in range(1, 6)]  # P001~P005
tests = [
    ("CRP", "mg/L", 0, 25),        # 임의 범위
    ("WBC", "10^3/uL", 3, 20),
    ("TEMP", "C", 36.0, 39.5),
]

rows = []
start = datetime(2026, 2, 20, 9, 0, 0)

visit_counter = 1000
for p in patients:
    # 환자당 2번 방문
    for _ in range(2):
        visit_counter += 1
        visit_id = f"V{visit_counter}"
        base_time = start + timedelta(days=random.randint(0, 10), hours=random.randint(0, 6))
        # 방문당 검사 3~6개
        for _ in range(random.randint(3, 6)):
            test_name, unit, lo, hi = random.choice(tests)
            collected_at = base_time + timedelta(minutes=random.randint(0, 240))
            if test_name == "TEMP":
                value = round(random.uniform(lo, hi), 1)
            else:
                value = round(random.uniform(lo, hi), 2)

            rows.append({
                "patient_id": p,
                "visit_id": visit_id,
                "collected_at": collected_at.strftime("%Y-%m-%d %H:%M:%S"),
                "test_name": test_name,
                "value": value,
                "unit": unit
            })

df = pd.DataFrame(rows).sort_values(["patient_id", "visit_id", "collected_at"])
df.to_csv(out_path, index=False, encoding="utf-8")
print(f"✅ wrote {len(df)} rows to {out_path}")
print(df.head(10).to_string(index=False))
```
